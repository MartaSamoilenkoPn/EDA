Analysis of cost of living in different cities

```python
import pandas as pd

def cost_of_living():
    data_frame = pd.read_csv('cost-of-living-2016.csv', index_col= 0, skiprows=0)
    list_country=['Netherlands','Germany','France','Switzerland','Austria','Belgium','Poland','Romania','Slovakia','Hungary','Ukraine','Czech Republic','Bulgaria','Sweden','Norway','Finland','Estonia','Latvia','Lithuania','United Kingdom','Denmark','Spain','Portugal','Italy','Croatia','Greece','Albania','Serbia','Montenegro']
    data_frame=data_frame.loc[data_frame.Country.isin(list_country)]
    return data_frame

cost_of_living()
```

Groupping by countries and counting an average value of Cost Of Living Index

```python
def group_by_country(data_frame):
    list_country=['Netherlands','Germany','France','Switzerland','Austria','Belgium','Poland','Romania','Slovakia','Hungary','Ukraine','Czech Republic','Bulgaria','Sweden','Norway','Finland','Estonia','Latvia','Lithuania','United Kingdom','Denmark','Spain','Portugal','Italy','Croatia','Greece','Albania','Serbia','Montenegro']
    data_frame=data_frame.loc[data_frame.Country.isin(list_country)]
    return data_frame.groupby(by = ['Country'], axis = 0).mean().sort_values('Cost.of.Living.Index', ascending=False).head(30)

group_by_country(cost_of_living())
```

Graphic of Rent, Living and Living+Rent Indexes.
Here we can see how rent affects the index.

```python
import matplotlib.pyplot as plt
df = group_by_country(cost_of_living()).sort_values('Cost.of.Living.Plus.Rent.Index', ascending=False)
df = df.reset_index(drop = False)
df.plot.bar(x = 'Country', y=['Rent.Index', 'Cost.of.Living.Plus.Rent.Index', 'Cost.of.Living.Index'])
# plt.show()
```
